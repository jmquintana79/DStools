# Global Model Interpretability with PyCaret: CLASSIFICATION

### References:
- [[Docs-Pycaret] Analyze](https://pycaret.gitbook.io/docs/get-started/functions/analyze)

```python
from pycaret.classification import *
from pycaret.datasets import get_data
```

## load data

```python
data = get_data('iris')
data.shape
```

### model

```python
sm = setup(
    data = data,
    target = "species",
    train_size = 0.8,
    test_data = None,
    data_split_shuffle = True,
    fold_strategy = "kfold",
    fold = 5,
    fold_shuffle = True,
    n_jobs = -1,
    verbose = True
)

# creating a model
clf = create_model('rf')
```

### model interpretability

```python
plot_model(clf, use_train_data=False, plot='auc')
```

```python
plot_model(clf, use_train_data=False,plot = 'confusion_matrix', plot_kwargs = {'percent' : True})
```

```python
plot_model(clf, use_train_data=False, plot='error')
```

```python
plot_model(clf, use_train_data=False, plot='class_report')
```

```python
plot_model(clf, use_train_data=False, plot='learning')
```

```python
plot_model(clf, use_train_data=False, plot='vc')
```

```python
plot_model(clf, use_train_data=False, plot='feature')
```

```python
interpret_model(clf, plot = 'pfi')
```

```python
interpret_model(clf, plot = 'pdp', feature = 'petal_length')
```

```python
# interpret model
interpret_model(clf, use_train_data = False)
```

```python
# feature = target feature to be analized in x-axis. In y-axis I think is selected automatically with the most correlated 
interpret_model(clf, plot = "correlation", feature = "petal_length", use_train_data = False)
```

```python

interpret_model(clf, plot = 'reason', use_train_data = False)
```
