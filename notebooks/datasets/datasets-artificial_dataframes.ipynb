# Artificial Pandas Dataframes

Creation of actificial Pandas dataframe for developing and testing.

### Reference:

- [How to create dataframe for testing?](https://kanoki.org/2019/11/18/how-to-create-dataframe-for-testing/)

```python
%matplotlib inline
import pandas as pd
import pandas.util.testing as tm
```

```python
# set dimmensions 
tm.N, tm.K = 10, 7
```

#### TIMESERIES DATAFRAME

```python
# it creates the time-series dataframe and is scaled x 100
st = pd.util.testing.makeTimeDataFrame() * 100
st.head()
```

#### RANDOM DATAFRAME

```python
# the index is a group of alpha-numeric character
df= pd.util.testing.makeDataFrame()
df.head()
```

#### WITH MISSING VALUES DATAFRAME

```python
# it creates dataframe with missing values in it
df= pd.util.testing.makeMissingDataframe()
df.head()
```

#### MIXED DATAFRAME

```python
# it creates a mixed dataframe containing Categorical, date-time and Continuous columns
df= pd.util.testing.makeMixedDataFrame()
df.head()
```

#### PERIODIC TIMESERIES DATAFRAME

```python
# it creates time-series dataframe with periodical data
df=pd.util.testing.makePeriodFrame()
df.head()
```
