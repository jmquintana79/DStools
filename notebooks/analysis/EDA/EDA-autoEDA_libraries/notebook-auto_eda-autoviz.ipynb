```python
%matplotlib inline
import seaborn
from autoviz import AutoViz_Class
from sklearn.datasets import load_iris
import pandas as pd
```

### dataset 1: Classification

```python
data = seaborn.load_dataset('titanic')
```

```python
AV = AutoViz_Class()
dft = AV.AutoViz(
    "",
    sep=",",
    depVar="survived",
    dfte=data,
    header=0,
    verbose=1, # = 2 in local
    lowess=False,
    chart_format="png",
    max_rows_analyzed=150000,
    max_cols_analyzed=30,
    save_plot_dir=None
)
```

### dataset 2: Classification (Iris)

```python
# load dataset
dataset = load_iris()
dataset.keys()
# dataset to df
data = pd.DataFrame(dataset.data, columns = dataset.feature_names)
data['class'] = dataset.target
dclass = dict()
for i, ic in enumerate(dataset.target_names):
    dclass[i] = ic
data['class'] = data['class'].map(dclass)
data.info()
```

```python
AV = AutoViz_Class()
dft = AV.AutoViz(
    "",
    sep=",",
    depVar="class",
    dfte=data,
    header=0,
    verbose=1, # = 2 in local
    lowess=False,
    chart_format="png",
    max_rows_analyzed=150000,
    max_cols_analyzed=30,
    save_plot_dir=None
)
```
