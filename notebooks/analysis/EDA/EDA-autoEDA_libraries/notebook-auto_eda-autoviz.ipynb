# Review Auto-EDA library: AUTOVIZ

No esta mal, incluso permite leer directamente el fichero. 
Lo malo es que en "bokeh", no te permite update el plot al seleccionar otra variable.

Para instalar: `pip install autoviz`

#### References

- [Docs PyPi Autoviz](https://pypi.org/project/autoviz/)
- [GitHub Autoviz](https://github.com/AutoViML/AutoViz)


```python
import seaborn
from autoviz import AutoViz_Class
```

### dataset 1: Classification

```python
data = seaborn.load_dataset('titanic')
```

```python
AV = AutoViz_Class()
dft = AV.AutoViz(
    "",
    sep=",",
    depVar="survived",
    dfte=data,
    header=0,
    verbose=1,
    lowess=False,
    chart_format="bokeh",
    max_rows_analyzed=150000,
    max_cols_analyzed=30,
    save_plot_dir=None
)
```
