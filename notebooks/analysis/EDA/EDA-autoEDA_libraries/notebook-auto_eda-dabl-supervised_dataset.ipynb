# Review Auto-EDA library: DABL

Es recomendable pero solo se usa para Supervised datasets, es decir, con X e y, tanto para problemas de regression como de classification.

No creo que sea recomendable ser usado para datasets con muchas columnas.

Para instalar: `pip install dabl`

#### References

- [Docs DABL](https://dabl.github.io/stable/index.html)

```python
import dabl
import seaborn
```

### dataset 1: Classification

```python
data = seaborn.load_dataset('titanic')
data_clean = dabl.clean(data, verbose=1)
types = dabl.detect_types(data_clean)
print(types) 
```

```python
dabl.plot?
```

```python
dabl.plot(data_clean, target_col = "survived")
```
