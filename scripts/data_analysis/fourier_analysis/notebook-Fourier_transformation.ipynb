# Fourier Transformation

The Fourier transform (FT) decomposes a function of time (a signal) into the frequencies. 


```python
%matplotlib inline
from IPython.display import Image
from IPython.core.display import HTML 
import numpy as np
from scipy.fftpack import fft, fftfreq
```

```python
Image(url= "https://slideplayer.com/slide/5661859/6/images/43/Link+between+Continuous+and+Discrete.jpg")
```

### signal definition

```python
# build time vector
Fs = 150.0;  # sampling rate
Ts = 1.0/Fs; # sampling interval
t = np.arange(0,1,Ts) # time vector
# build signal vector
ff = 5;   # frequency of the signal
y = np.sin(2*np.pi*ff*t)
```

## transformation using ***Numpy***

```python
# build frequency vector
n = len(y)
timestep = t[1] - t[0]
frq = np.fft.fftfreq(n, d=timestep)
```

```python
import matplotlib.pyplot as plt

## transformation
Y1 = np.fft.fft(y)/n # fft computing and normalization

## plot
fig, ax = plt.subplots(2, 1)
ax[0].plot(t,y)
ax[0].set_xlabel('Time')
ax[0].set_ylabel('Amplitude')
ax[1].plot(frq,abs(Y1),'r') # plotting the spectrum
ax[1].set_xlabel('Freq (Hz)')
ax[1].set_ylabel('|Y(freq)|')
plt.show()
```

```python
# frequency of the signal
fr = list(set(np.abs(frq[np.where(abs(Y1)>np.percentile(abs(Y1),99))[0]])))
print('frequency of the signal: %s'%fr)
```

#### Consequences:
- Effectively, the frequency of the signal is just 5, that is, in the whole signal, periodic signal is repeated 5 times.

## transformation using **Scipy**
- source: [Fourier Transformation](https://docs.scipy.org/doc/scipy/reference/tutorial/fftpack.html).

```python
# build frequency vector
n = len(y)
timestep = t[1] - t[0]
frq = fftfreq(n, d=timestep)
```

```python
import matplotlib.pyplot as plt

## transformation
Y2 = fft(y)
#Y2 = Y2[range(int(n/2))]

## plot
fig, ax = plt.subplots(2, 1)
ax[0].plot(t,y)
ax[0].set_xlabel('Time')
ax[0].set_ylabel('Amplitude')
ax[1].plot(frq,abs(Y2),'r') # plotting the spectrum
ax[1].set_xlabel('Freq (Hz)')
ax[1].set_ylabel('|Y(freq)|')
plt.show()
```

```python
# frequency of the signal
fr = list(set(np.abs(frq[np.where(abs(Y2)>np.percentile(abs(Y2),99))[0]])))
print('frequency of the signal: %s'%fr)
```
